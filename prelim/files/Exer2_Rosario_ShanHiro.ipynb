## Applying to dataset

This is a helper function to remove stopword from the article column. In the start of for loop, the article.split() simply tokenize each stopwords by whitespace. The if-else statements are the same as the final_test variable above. (Skip empty, linebreaks, or stopwords then print)

```python
def remove_stopwords(article,stopwords):
    temp_fnl = []
    for word in article.split():
        if word =='' or '\r\n' in word or word in stopwords:
            None
        else:
            temp_fnl.append(word)
    return ' '.join(temp_fnl)
```

We use the function created to remove the stopwords of the test_case variable then print it as we can see  here. Note that it is only an example, the next codeblock applies the function to the whole dataset

It is important to remove special characters first from the test_case to ensure that we are removing the correct ones. Here we imported the regex for removing.

```python
import re

test_case = re.sub("[^a-zA-Z0-9\s]", '', test_case)
print(test_case)

```

Then we apply the same logic to the whole dataset, in addition to removing URL's because we are only detecting whether the message is a spam message or a real one. The URL, i believe is outside of scope as it is only useful for phishing detection, and web scraping features

```python
def count_removals(text):
    url_matches = re.findall(r'http\S+|www\.\S+', text)
    special_char_matches = re.findall(r'[^a-zA-Z0-9\s]', text)
    return pd.Series([len(url_matches), len(special_char_matches)])

df[['URLs_Removed', 'SpecialChars_Removed']] = df['Raw_Message'].apply(count_removals)

# Step 3: Now apply your original cleaning (unchanged)
df['Message'] = df['Message'].apply(lambda x: re.sub("[^a-zA-Z0-9\s]", '', x))  # remove special chars
df['Message'] = df['Message'].apply(lambda x: re.sub(r'http\S+|www.\S+', '', x))  # remove URLs

# Step 4: View results
df[['Raw_Message', 'Message', 'URLs_Removed', 'SpecialChars_Removed']].head(10)
```

We import the required library for the available english stopwords and manually added singaporean stopwords in a dictionary format as the messages are composed of mixed singapore and english in their context. Then we added singapore_stopwords to the variable stop_words which contains english stopwords

```python
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

nltk.download('stopwords')
stop_words = stopwords.words('english')

singapore_stopwords = {
    'lah', 'lor', 'leh', 'liao', 'meh', 'mah', 'hor', 'ah', 'eh', 'loh',
    'ar', 'deh', 'siah', 'liaoz', 'siol', 'lehz', 'la', 'haiz', 'wa',
    'kan', 'den', 'oso', 'wan', 'wun', 'alr', 'liaoz', 'liao', 'leh', 'lor','lar'
}

stop_words.extend(singapore_stopwords)
```

These are the stopword removal using for loop and if else similar to the instructions provided, whereas if the word is an empty space, a new line or is a word in the stop_words dictionary, do not remove it then simply append white space using .join for better reading.

```python
final_test = []
for word in test_case.split():
    if word == '' or '\r\n' in word or word in stop_words:
        None
    else:
        final_test.append(word)

print(final_test)
' '.join(final_test)
```

A helper function, a reusable one to remove stopwords easily in coding including the one above. Then we apply the function to the  entire 'Message' Column in the dataset then preview the first 10 rows

```python
def remove_stopwords(article, stopwords):
    temp_fnl = []
    removed_count = 0

    for word in article.split():
        clean_word = word.strip().lower()
        if clean_word == '' or clean_word in stopwords:
            removed_count += 1
        else:
            temp_fnl.append(word)

    cleaned_text = ' '.join(temp_fnl)
    return pd.Series([cleaned_text, removed_count])
df[['Message_Cleaned', 'Stopwords_Removed']] = df['Message'].apply(
    lambda x: remove_stopwords(x, stop_words)
)

df.head(10)
```

We create a reusable function again for normalization this time. Then apply the function to the previous tokenized message then show the results.

```python
def normalize_slang(tokens, slang_dict):
    return [slang_dict.get(word, word) for word in tokens]

df['tokenized_message'] = df['Message_Cleaned'].apply(word_tokenize)
df['normalized_message'] = df['tokenized_message'].apply(lambda tokens: normalize_slang(tokens, slang_dict))

df[['tokenized_message', 'normalized_message']].head(10)
```

Add Total sum for existing columns, then
Count words removed from the initial column of the dataset to the cleaned column, then  Print totals

```python
# Total sum for existing columns
total_urls_removed = df['URLs_Removed'].sum()
total_specials_removed = df['SpecialChars_Removed'].sum()
total_stopwords_removed = df['Stopwords_Removed'].sum()

# Count words removed from Raw_Message to normalized_message
# Assuming normalized_message is a list of tokens (after stopword removal + slang normalization)
def count_words_removed(row):
    raw_count = len(str(row['Raw_Message']).split())
    norm_count = len(row['normalized_message'])  # already tokenized list
    return raw_count - norm_count

df['Words_Removed'] = df.apply(count_words_removed, axis=1)
total_words_removed = df['Words_Removed'].sum()

# Print totals
print("Total URLs Removed:", total_urls_removed)
print("Total Special Characters Removed:", total_specials_removed)
print("Total Stopwords Removed:", total_stopwords_removed)
print("Total Words Removed (Raw → Normalized):", total_words_removed)
```
